```python
import pandas as pd
import numpy as np
```

```python
player2019 = pd.read_csv("phs_2019_stage_1.csv")
```

```python
player2019.head()
```

```python
herotime = player2019.loc[player2019["stat_name"] == "Time Played"]
herotime = herotime.loc[player2019["hero"] != "All Heroes"]
herotime.head()
```

```python
heroesonly = herotime[["map_name","hero","stat_amount"]]
heroesonly = heroesonly.rename(columns={'stat_amount': 'Time Played'})
heroesonly = heroesonly.groupby(["map_name","hero"]).agg("sum")
heroesonly
```

```python
heroesonly["Most Played"] = heroesonly.groupby(["map_name"])["Time Played"].rank("dense", ascending = False)
heroesonly = heroesonly.sort_values(["map_name", "Most Played"],ascending = True)
heroesonly = heroesonly.loc[heroesonly["Most Played"]<= 6]
heroesonly
```
