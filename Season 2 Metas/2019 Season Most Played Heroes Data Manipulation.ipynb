```python
import pandas as pd
import numpy as np
```

```python
S1player2019 = pd.read_csv("C:/Users/William Wu/Documents/Overwatch League Stats/Raw Data/phs_2019_stage_1.csv")
S2player2019 = pd.read_csv("C:/Users/William Wu/Documents/Overwatch League Stats/Raw Data/phs_2019_stage_2.csv")
S3player2019 = pd.read_csv("C:/Users/William Wu/Documents/Overwatch League Stats/Raw Data/phs_2019_stage_3.csv")
S4player2019 = pd.read_csv("C:/Users/William Wu/Documents/Overwatch League Stats/Raw Data/phs_2019_stage_4.csv")
SPplayer2019 = pd.read_csv("C:/Users/William Wu/Documents/Overwatch League Stats/Raw Data/phs_2019_playoffs.csv")
```

```python
S1player2019.head()
```

```python
S1player2019["stage"] = "Stage 1"
S2player2019["stage"] = "Stage 2"
S3player2019["stage"] = "Stage 3"
S4player2019["stage"] = "Stage 4"
SPplayer2019["stage"] = "Playoffs"
SPplayer2019.head()
```

```python
player2019 = S1player2019.append(S2player2019, ignore_index = True).append(S3player2019, ignore_index = True).append(S4player2019, ignore_index = True).append(SPplayer2019, ignore_index = True)
player2019.head()
```

```python
herotime = player2019.loc[player2019["stat_name"] == "Time Played"]
herotime = herotime.loc[player2019["hero"] != "All Heroes"]
herotime.head()
```

```python
heroesonly = herotime[["map_name","hero","stage","stat_amount"]]
heroesonly = heroesonly.rename(columns={'stat_amount': 'Time Played'})
heroesonly = heroesonly.groupby(["stage","map_name","hero"]).agg("sum")
heroesonly
```

```python
heroesonly["Most Played"] = heroesonly.groupby(["stage","map_name"])["Time Played"].rank("dense", ascending = False)
heroesonly = heroesonly.sort_values(["stage","map_name", "Most Played"],ascending = True)
#heroesonly = heroesonly.loc[heroesonly["Most Played"]<= 6]
heroesonly
```

```python
heroesonly.to_csv("2019playtime.csv")
```
